### Calculate chelsea wage and analyze correlation between overall and wage

```python
CHE=df.loc[df['Club'] == "Chelsea"]
CHE1=CHE[['Name', 'Age','Position','Overall','Wage','Release Clause']]
CHE1
```

```python
CHE1['Wage'] = CHE1['Wage'].replace({'€': '', ' ': '', 'M': 'e+06', 'K': 'e+03'}, regex=True).astype(float).astype(int)
CHE1['Release Clause'] = CHE1['Release Clause'].replace({'€': '', ' ': '', 'M': 'e+06', 'K': 'e+03'}, regex=True).astype(float)
CHE1
```

```python
chelsea_wage = CHE1["Wage"].sum()
print("The Sum of Chelsea player's wage is €",chelsea_wage)
```

```python
plt.figure(figsize=(20, 10))
sns.regplot(x="Overall", y="Wage", data = CHE1)
```
